```python
import pandas as pd

CC = pd.read_csv("./CyberChair.csv")
CACP = pd.read_csv("./Collected_ACP_Sentences.csv")
IBM = pd.read_csv("./IBM_Course_Management.csv")
i_ACRE = pd.read_csv("./iTrust_forACRE.csv")
i_T2P = pd.read_csv("./iTrust_Text2Policy.csv")

data_list =[CC,CACP,IBM,i_ACRE,i_T2P]
data_name = ["CyberChair","Collected_ACP","IBM_Course_Management","Itrust_ACRE","Text2Policy"]
```

```python
CC.type.value_counts()
```

```python
count = 0
for dataset in data_list:
        print(data_name[count])
        dataset['XiaoACP'] = dataset['XiaoACP'].fillna("False")
        print(dataset.AccessControl.value_counts())
        print(dataset.XiaoACP.value_counts())
        count += 1
```

```python
count = 0
new = []
for dataset in data_list:
        dataset_re1 = dataset.replace([True,"True","TRUE"],1)
        dataset_re2 = dataset_re1.replace([False,"FALSE","false","False"],0)
        new.append(dataset_re2)
        #dataset_re2.to_csv(name+"_convert2.csv")
```

```python
new
```

```python
count = 0
for dataset in new:
        #for index in dataset:
        dataset['XiaoACP'] = dataset['XiaoACP'].fillna("false")
        AC = dataset.AccessControl
        X_AC =dataset.XiaoACP
        diff = AC[AC!=X_AC]
        #print(CC.Sentence[diff.index])
        diff_sen = dataset.Sentence[diff.index]
        diff_AC = dataset.AccessControl[diff.index]
        diff_X_AC = dataset.XiaoACP[diff.index]
        #diff_verb = dataset.verb[diff.index]
        merge = pd.DataFrame({ "type":dataset.type[diff.index],"AccessControl":diff_AC,"Xiao_AC":diff_X_AC,"Sentence":diff_sen})
        merge.info()
        merge.to_csv(data_name[count]+"_diff.csv")
        count +=1
```
